```python
product_data = [
    (101, "Laptop", 65000),
    (102, "Mobile", 25000),
    (103, "Keyboard", 1500),
    (104, "Mouse", 800),
    (105, "Monitor", 12000)
]
```

```python
columns = ["product_id", "product_name", "price"]

df = spark.createDataFrame(product_data, columns)
```

```python
df.show()
```

```python
df.createOrReplaceTempView("products")
```

```python


avg_price = spark.sql("""
SELECT AVG(price) AS average_price
FROM products
""")
```

```python
avg_price.show()
```
