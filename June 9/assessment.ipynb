```python
# Transaction Statistics
total_transactions = rdd.count()
unique_transactions = rdd.distinct().count()

transaction_counts = rdd.countByValue()

most_common = max(
    transaction_counts.items(),
    key=lambda x: x[1]
)

least_common = min(
    transaction_counts.items(),
    key=lambda x: x[1]
)

# Amount Statistics
total_amount = amount_rdd.reduce(
    lambda a, b: a + b
)

average_amount = total_amount / amount_rdd.count()

max_amount = amount_rdd.reduce(
    lambda a, b: a if a > b else b
)

min_amount = amount_rdd.reduce(
    lambda a, b: a if a < b else b
)

high_value_count = amount_rdd.filter(
    lambda x: x > 20000
).count()

low_value_count = amount_rdd.filter(
    lambda x: x < 1000
).count()

# Top 3 Transaction Types
top_3 = sorted(
    transaction_counts.items(),
    key=lambda x: x[1],
    reverse=True
)[:3]

# Dashboard Output
print("=" * 50)
print("BANK TRANSACTION ANALYTICS DASHBOARD")
print("=" * 50)

print("\nTRANSACTION SUMMARY")
print("-" * 25)
print("Total Transactions :", total_transactions)
print("Unique Transaction Types :", unique_transactions)

print("\nMOST / LEAST FREQUENT")
print("-" * 25)
print("Most Frequent :", most_common)
print("Least Frequent :", least_common)

print("\nTRANSACTION FREQUENCIES")
print("-" * 25)

for transaction, count in sorted(
        transaction_counts.items(),
        key=lambda x: x[1],
        reverse=True):
    print(f"{transaction:<20} {count}")

print("\nTOP 3 TRANSACTION TYPES")
print("-" * 25)

for transaction, count in top_3:
    print(f"{transaction:<20} {count}")

print("\nAMOUNT ANALYTICS")
print("-" * 25)
print("Total Amount Processed :", total_amount)
print("Average Amount :", round(average_amount, 2))
print("Maximum Amount :", max_amount)
print("Minimum Amount :", min_amount)

print("\nVALUE SEGMENTATION")
print("-" * 25)
print("High Value Transactions (>20000) :", high_value_count)
print("Low Value Transactions (<1000) :", low_value_count)

print("\nHIGH VALUE AMOUNTS")
print("-" * 25)
print(
    amount_rdd.filter(
        lambda x: x > 20000
    ).collect()
)

print("\nLOW VALUE AMOUNTS")
print("-" * 25)
print(
    amount_rdd.filter(
        lambda x: x < 1000
    ).collect()
)

print("\nTOP 10 HIGHEST AMOUNTS")
print("-" * 25)
print(amount_rdd.top(10))

print("\nTOP 10 LOWEST AMOUNTS")
print("-" * 25)
print(amount_rdd.takeOrdered(10))

print("\nDASHBOARD GENERATED SUCCESSFULLY")
print("=" * 50)
```
