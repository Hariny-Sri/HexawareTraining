```python
unique_activities = rdd.distinct().sortBy(lambda x: x)

print("Unique Activities")
print("-----------------")

for activity in unique_activities.collect():
    print(activity)
```

```python
counts = rdd.countByValue()

print("Frequency Report")
print("----------------")

for activity, count in sorted(
        counts.items(),
        key=lambda x: x[1],
        reverse=True):
    print(activity, ":", count)
```
