```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

df_rev = df.withColumn(
    "revenue",
    col("quantity") * col("unit_price")
)
```

```python
w = Window.orderBy(desc("revenue"))

df_rev.withColumn(
    "row_number",
    row_number().over(w)
).show()
```

```python
df_rev.withColumn(
    "rank",
    rank().over(w)
).show()
```

```python
df_rev.withColumn(
    "dense_rank",
    dense_rank().over(w)
).show()
```

```python
df_rev.orderBy(
    desc("revenue")
).show(5)
```

```python
df_rev.orderBy(
    desc("revenue")
).show(1)
```

```python
w_city = Window.partitionBy("city").orderBy(desc("revenue"))

df_rev.withColumn(
    "rank",
    row_number().over(w_city)
).filter(
    col("rank") == 1
).show()
```

```python
w_cat = Window.partitionBy("category").orderBy(desc("revenue"))

df_rev.withColumn(
    "rank",
    row_number().over(w_cat)
).filter(
    col("rank") == 1
).show()
```

```python
w_run = Window.orderBy("order_id")

df_rev.withColumn(
    "running_revenue",
    sum("revenue").over(
        w_run.rowsBetween(
            Window.unboundedPreceding,
            Window.currentRow
        )
    )
).show()
```

```python
df_rev.withColumn(
    "previous_revenue",
    lag("revenue").over(
        Window.orderBy("order_id")
    )
).show()
```

```python
df_rev.withColumn(
    "next_revenue",
    lead("revenue").over(
        Window.orderBy("order_id")
    )
).show()
```

```python
df_rev.withColumn(
    "previous_revenue",
    lag("revenue").over(
        Window.orderBy("order_id")
    )
).withColumn(
    "comparison",
    when(
        col("revenue") > col("previous_revenue"),
        "Higher"
    ).when(
        col("revenue") < col("previous_revenue"),
        "Lower"
    ).otherwise("Same")
).show()
```

```python
df_rev.withColumn(
    "previous_revenue",
    lag("revenue").over(
        Window.orderBy("order_id")
    )
).withColumn(
    "revenue_difference",
    col("revenue") - col("previous_revenue")
).show()
```

```python
w_city = Window.partitionBy(
    "city"
).orderBy(
    desc("revenue")
)

df_rev.withColumn(
    "rank",
    row_number().over(w_city)
).filter(
    col("rank") <= 2
).show()
```

```python
w_cat = Window.partitionBy(
    "category"
).orderBy(
    desc("revenue")
)

df_rev.withColumn(
    "rank",
    row_number().over(w_cat)
).filter(
    col("rank") <= 3
).show()
```

```python
revenue_report = df_rev.withColumn(
    "row_number",
    row_number().over(w)
).withColumn(
    "rank",
    rank().over(w)
).withColumn(
    "dense_rank",
    dense_rank().over(w)
)

revenue_report.select(
    "order_id",
    "product",
    "city",
    "revenue",
    "row_number",
    "rank",
    "dense_rank"
).show()
```

```python
df = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("/Volumes/ey_data/default/myvolume/sales_data.csv")

df.show()
```

```python
from pyspark.sql.functions import col

df_calc = df.withColumn(
    "total_amount",
    col("quantity") * col("price")
).withColumn(
    "tax",
    col("total_amount") * 0.05
).withColumn(
    "final_amount",
    col("total_amount") + col("tax")
)

df_calc.show()
```

```python
completed_df = df_calc.filter(
    col("status") == "Completed"
)

completed_df.show()
```

```python
from pyspark.sql.functions import sum

completed_df.agg(
    sum("total_amount").alias("total_revenue")
).show()
```

```python
city_revenue_report = completed_df.groupBy(
    "city"
).agg(
    sum("total_amount").alias("total_revenue")
)

city_revenue_report.show()
```

```python
category_revenue_report = completed_df.groupBy(
    "category"
).agg(
    sum("total_amount").alias("total_revenue")
)

category_revenue_report.show()
```

```python
df.write.mode("overwrite").parquet(
    "/Volumes/ey_data/default/myvolume/sales_parquet"
)
```

```python
parquet_df = spark.read.parquet(
    "/Volumes/ey_data/default/myvolume/sales_parquet"
)

parquet_df.show()
```

```python
parquet_df.createOrReplaceTempView(
    "sales_report"
)
```
