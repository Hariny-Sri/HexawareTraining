```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Working with Files") \
    .getOrCreate()
```

```python
df = spark.read.option(
    "multiline",
    "true"
).json("hotels.json")

df.show(truncate=False)
```

```python
from pyspark.sql.functions import *
```

```python
df.show()
```

```python
df.select("hotel_name","city","rating").show()
```

```python
df.filter(col("category")=="Luxury").show()
```

```python
df.filter(col("rating") > 4.5).show()
```

```python
df.filter(col("rooms_available") > 15).show()
```

```python
df.filter(col("price_per_night") > 10000).show()
```

```python
df.filter(col("city").isin("Dubai", "London")).show()
```

```python
df.filter(col("contact.phone").isNull()).show()
```

```python
df.filter(col("contact.email").isNull()).show()
```

```python
df.select(
    "hotel_name",
    "contact.phone",
    "contact.email"
).show(truncate=False)
```

```python
df.filter(array_contains(col("amenities"), "wifi")).show()
```

```python
df.filter(array_contains(col("amenities"), "spa")).show(truncate=False)
```

```python
df.groupBy(col("city")).count().show()
```

```python
df.groupBy( "category").count().show()
```

```python
df.groupBy("category").agg(avg("rating").alias("avg_rating")).show()
```

```python
df.groupBy("city").agg(avg("price_per_night").alias("avg_price")).show()
```

```python
df.agg(max("price_per_night").alias("highest_price")).show()
```

```python
revenue_df = df.withColumn("total_potential_revenue",col("rooms_available") * col("price_per_night"))

revenue_df.show()
```

```python
df.orderBy(col("rating").desc()).show()
```

```python
flat_df = df.select(
    "hotel_id",
    "hotel_name",
    "city",
    "category",
    "rating",
    "rooms_available",
    "price_per_night",
    col("contact.phone").alias("phone"),
    col("contact.email").alias("email")
)
```
