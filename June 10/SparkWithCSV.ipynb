```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Working with Files") \
    .getOrCreate()
```

```python
df = spark.read.csv(
    "flights.csv",
    header=True,
    inferSchema=True
)
```

```python
df.show()
```

```python
df.show()
```

```python
df.select("passenger_name","from_city","to_city").show()
```

```python
from pyspark.sql.functions import *
```

```python
df.filter(df.status=="Confirmed").show()
```

```python
df.filter(df.status=="Cancelled").show()
```

```python
df.filter(df.status=="Pending").show()
```

```python
df.filter(df.ticket_price>20000).show()
```

```python
df.filter(col("to_city").isin("Dubai","Singapore","London")).show()
```

```python
df.count()
```

```python
df.groupBy("airline").count().show()
```

```python
df.groupBy("status").count().show()
```

```python
df.filter(col("status")=="Confirmed").agg(sum("ticket_price").alias("total_revenue")).show()
```

```python
df.groupBy("airline").agg(avg("ticket_price").alias("avg_price")).show()
```

```python
df.agg(max("ticket_price").alias("highest_price")).show()
```

```python
df.agg(min("ticket_price").alias("highest_price")).show()
```

```python
df.orderBy(col("ticket_price").desc()).show()
```

```python
df_tax=df.withColumn("tax",col("ticket_price")*0.05)
df_tax.show()
```

```python
df_final = df_tax.withColumn( "final_price",col("ticket_price") + col("tax"))
df_final.show()
```

```python
df_category = df.withColumn(
    "price_category",
    when(col("ticket_price") >= 30000, "Premium")
    .when(col("ticket_price") >= 10000, "Standard")
    .otherwise("Budget"))

df_category.show()
```

```python
df_category.groupBy("price_category").count().show()
```

```python
df.filter(
    col("status") == "Confirmed"
).write.mode("overwrite").parquet(
    "confirmed_flights.parquet"
)
```
