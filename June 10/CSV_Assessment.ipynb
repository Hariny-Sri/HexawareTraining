```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Bookings Analysis") \
    .getOrCreate()
```

```python
from pyspark.sql.functions import *
```

```python
quality_df = df.withColumn(
    "data_quality_status",
    when(
        col("customer_name").isNull() |
        col("service_type").isNull() |
        col("booking_amount").isNull() |
        col("city").isNull() |
        col("provider").isNull(),
        "Incomplete"
    ).otherwise("Complete")
)

quality_df.show()
```

```python
quality_df.groupBy("data_quality_status").count().show()
```

```python
quality_df.filter(col("data_quality_status")=="Complete").show()
```

```python
quality_df.filter(col("data_quality_status")=="Incomplete").show()
```

```python
tax_df=df.na.fill({"booking_amount":0})
tax_df=tax_df.withColumn("tax",col("booking_amount")*0.05)
tax_df.show()
```

```python
final_amount=tax_df.withColumn("Final Amount",col("booking_amount")+col("tax"))
final_amount.show()
```

```python
df.filter(col("booking_status")=="Confirmed").agg(sum("booking_amount").alias("revenue")).show()
```

```python
service_df = df.na.fill({"service_type": "Unknown"})
service_df.groupBy("service_type").count().show()
```

```python
booking_df = df.na.fill({"city": "Unknown"})
booking_df.groupBy("city").count().show()
```

```python
df.na.fill({"booking_amount": 0}).agg(avg("booking_amount").alias("avg_amount")).show()
```

```python
df.na.drop(subset=["booking_amount"]).agg(avg("booking_amount").alias("avg_amount")).show()
```

```python
avg_with_zero = df.na.fill(
    {"booking_amount": 0}
).agg(
    avg("booking_amount").alias("avg_with_zero")
)

avg_after_drop = df.na.drop(
    subset=["booking_amount"]
).agg(
    avg("booking_amount").alias("avg_after_drop")
)

avg_with_zero.show()
avg_after_drop.show()
```
