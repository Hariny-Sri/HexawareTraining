```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("CustomerJSON") \
    .getOrCreate()
df = spark.read.option(
    "multiline",
    "true"
).json("Customers.json")

df.show(truncate=False)
df.printSchema()
```

```python
from pyspark.sql.functions import *

flat_customers_df = df.select(
    "customer_id",
    "name",
    "city",
    "membership",
    col("contact.phone").alias("phone"),
    col("contact.email").alias("email"),
    col("preferences.preferred_service").alias("preferred_service"),
    col("preferences.budget_range").alias("budget_range")
)

flat_customers_df.show()
```

```python
flat_customers_df.na.fill({"phone": "Not Provided"})
```

```python
flat_customers_df.na.fill({
    "preferred_service": "Not Selected"
}).show()
```

```python
clean_df = flat_customers_df.na.fill({
    "budget_range": "Unknown"
})

clean_df.show()
```

```python
quality_df = df.withColumn(
    "customer_quality_status",
    when(
        col("city").isNull() |
        col("contact.phone").isNull() |
        col("contact.email").isNull() |
        col("membership").isNull() |
        col("preferences.preferred_service").isNull(),
        "Incomplete"
    ).otherwise("Complete")
)

quality_df.show()
```

```python
quality_df.groupBy(
    "customer_quality_status"
).count().show()
```

```python
quality_df.filter(
    col("customer_quality_status") == "Complete"
).show()
```

```python
quality_df.filter(
    col("customer_quality_status") == "Incomplete"
).show()
```

```python
clean_df.groupBy(
    "membership"
).count().show()
```

```python
clean_df.groupBy(
    "preferred_service"
).count().show()
```

```python
clean_df.write.mode("overwrite"
).option("header", True).csv("clean_customers.csv")
```

```python
print("Original Count:",
      df.count())

print("Clean Count:",
      clean_df.count())
```
