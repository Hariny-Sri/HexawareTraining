```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *
```

```python
product_revenue_df = sales_product_df.groupBy(
    "product_name"
).agg(
    sum("sale_amount").alias("revenue")
)

window_spec = Window.orderBy(desc("revenue"))

product_revenue_df.withColumn(
    "rank",
    rank().over(window_spec)
).show()
```

```python
store_revenue_df = sales_store_df.groupBy(
    "store_name"
).agg(
    sum("sale_amount").alias("revenue")
)

window_spec = Window.orderBy(desc("revenue"))

store_revenue_df.withColumn(
    "rank",
    rank().over(window_spec)
).show()
```

```python
product_category_df = sales_product_df.groupBy(
    "category",
    "product_name"
).agg(
    sum("sale_amount").alias("revenue")
)

window_spec = Window.partitionBy(
    "category"
).orderBy(
    desc("revenue")
)

product_category_df.withColumn(
    "rank",
    rank().over(window_spec)
).show()
```

```python
top_product_df = product_category_df.withColumn(
    "rank",
    rank().over(window_spec)
)

top_product_df.filter(
    col("rank") == 1
).show()
```

```python
top_product_df.filter(
    col("rank") <= 3
).show()
```

```python
store_state_df = sales_store_df.groupBy(
    "state",
    "store_name"
).agg(
    sum("sale_amount").alias("revenue")
)

window_spec = Window.partitionBy(
    "state"
).orderBy(
    desc("revenue")
)

store_state_df.withColumn(
    "rank",
    rank().over(window_spec)
).filter(
    col("rank") == 1
).show()
```

```python
window_spec = Window.orderBy("sale_date")

sales_df.withColumn(
    "running_revenue",
    sum("sale_amount").over(window_spec)
).show()
```

```python
window_spec = Window.orderBy("sale_date")

sales_df.withColumn(
    "previous_sale",
    lag("sale_amount",1).over(window_spec)
).show()
```

```python
window_spec = Window.orderBy("sale_date")

sales_df.withColumn(
    "next_sale",
    lead("sale_amount",1).over(window_spec)
).show()
```

```python
window_spec = Window.partitionBy(
    "product_id"
).orderBy(
    "sale_date"
)

sales_growth_df = sales_df.withColumn(
    "previous_sale",
    lag("sale_amount",1).over(window_spec)
)

sales_growth_df.filter(
    col("sale_amount") > col("previous_sale")
).show()
```
